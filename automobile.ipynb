### It seems there are no improper values in Symboling column

```python
df.iloc[:,0].unique()
```

### there are ? as values in column normalized-losses

```python
df.iloc[:,1].unique()
```

```python
df = df.replace("?", np.nan)
df.head(10)
```

### Since we have Nan values in Price column as well, we dont need them because its a target column(dependent variable)

```python
df[["price"]].head(11)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df[["price"]].head(11)
```

```python
df.info()
```

### make is just a brand name for the automobiles

```python
df["make"].unique()
```

### bore Nan values can be replaced with mean() of the variable

```python
df.loc[54:60, ["bore","stroke"]]
```

```python
df["bore"] = pd.to_numeric(df["bore"])
df["stroke"] = pd.to_numeric(df["stroke"])
df.loc[df["bore"].isnull(), "bore"] = df["bore"].mean()
df.loc[df["stroke"].isnull(), "stroke"] = df["stroke"].mean()
df.loc[54:60,["bore","stroke"]]
```

```python
df[["compression-ratio","horsepower","peak-rpm"]].info()
```

### horsepower and peak-rpm will have Nan value replaced with mode of the respective variable

```python
df.loc[128:132, ["horsepower","peak-rpm"]]
```

```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
df["peak-rpm"] = pd.to_numeric(df["peak-rpm"], errors="coerce")
df.loc[df["horsepower"].isnull(), "horsepower"] = round(df["horsepower"].mean(),1)
df.loc[df["peak-rpm"].isnull(), "peak-rpm"] = round(df["peak-rpm"].mean(),1)
```

```python
df.loc[128:132, ["horsepower","peak-rpm"]]
```

### city/highway-mpg means miles per gallons, we'll convert it to kilometer/Liter

```python
df[["city-mpg","highway-mpg"]].head(10)
```

```python
df["city-mpg"] = round(df["city-mpg"]*0.425143707, 2)
df["highway-mpg"] = round(df["highway-mpg"]*0.425143707, 2)
df.rename(columns={"city-mpg":"city-km/l", "highway-mpg":"highway-km/l"}, inplace=True)
df[["city-km/l","highway-km/l"]].head(10)
```

### converting the price datatype to float

```python
df["price"].head()
```

```python
df["price"] = df["price"].astype("float64")
df["price"].head()
```

```python
df.iloc[:,-4:].head(10)
```

```python
step = int((max(df["price"]) - min(df["price"]))/3)
step
```

```python
bins = range(int(min(df["price"])), int(max(df["price"])), step)
print(bins)
```

```python
df["budget"] = pd.cut(df["price"], bins, labels=["low", "medium", "high"])
df.iloc[:,-4:].head(10)
```

```python
df["num-of-doors"].value_counts()
```

```python
df["num-of-doors"] = df["num-of-doors"].replace(["two","four",np.nan], [2,4,4])
df.loc[[26,27,31,62,63],["num-of-doors"]]
```

```python
df.loc[df["budget"].isnull(), "budget"] = df["budget"].mode()
```
